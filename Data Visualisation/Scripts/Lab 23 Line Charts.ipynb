### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Remove NaN values from the 'ConvertedCompYearly' column
df_cleaned = df.dropna(subset=['ConvertedCompYearly'])

# Print unique values for age groups and cleaned ConvertedCompYearly
print("Unique Age:")
print(df_cleaned['Age'].unique())

print("\nUnique Yearly Compensation (ConvertedCompYearly) without NaN:")
print(df_cleaned['ConvertedCompYearly'].unique())
```

```python
## Write your code here

# Define the order of age groups
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old', 
             '45-54 years old', '55-64 years old', '65 years or older', 'Prefer not to say']

# Convert the 'Age' column to a categorical type with the specified order
df_cleaned['Age'] = pd.Categorical(df_cleaned['Age'], categories=age_order, ordered=True)

# Group by 'Age' and calculate the median 'ConvertedCompYearly'
age_median_compensation = df_cleaned.groupby('Age')['ConvertedCompYearly'].median()

# Plotting the line chart
plt.figure(figsize=(10, 6))
age_median_compensation.plot(kind='line', marker='o')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here

# Filter the data to focus on age groups between 25 and 45
filtered_age_groups = ['25-34 years old', '35-44 years old']
df_filtered = df_cleaned[df_cleaned['Age'].isin(filtered_age_groups)]

# Group by 'Age' and calculate the median 'ConvertedCompYearly'
age_median_compensation_filtered = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plotting the line chart for the filtered age groups
plt.figure(figsize=(10, 6))
age_median_compensation_filtered.plot(kind='line', marker='o')
plt.title('Median Yearly Compensation for Ages 25-45')
plt.xlabel('Age Group')
plt.ylabel('Median Yearly Compensation')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here

# Define the work experience ranges
bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 60, 100]
labels = ['1-5 years', '6-10 years', '11-15 years', '16-20 years', '21-25 years', 
          '26-30 years', '31-35 years', '36-40 years', '41-45 years', '46-50 years', 
          '51-60 years', '60+ years']

# Create a new column in the DataFrame for the categorized work experience
df_cleaned['WorkExpRange'] = pd.cut(df_cleaned['WorkExp'], bins=bins, labels=labels, right=False)

# Group by the 'WorkExpRange' and calculate the mean of 'JobSatPoints_6'
workexp_grouped = df_cleaned.groupby('WorkExpRange')['JobSatPoints_6'].mean()

# Plotting the line chart
plt.figure(figsize=(10, 6))
workexp_grouped.plot(kind='line', marker='o', linestyle='-', color='b')
plt.title('Average Job Satisfaction by Work Experience Range')
plt.xlabel('Work Experience Range')
plt.ylabel('Average Job Satisfaction')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
