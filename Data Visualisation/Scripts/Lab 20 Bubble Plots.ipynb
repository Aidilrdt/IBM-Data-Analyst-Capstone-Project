### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here

# Clean the data by removing NaN values in the relevant columns
df_cleaned_bubble = df.dropna(subset=['Age', 'SOPartFreq', 'JobSat'])

# Set a scaling factor to map the JobSat range to bubble sizes
bubble_size = df_cleaned_bubble['JobSat']

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=df_cleaned_bubble,
                x='Age', 
                y='SOPartFreq', 
                size=bubble_size, 
                hue='JobSat',  # Color the points by job satisfaction
                sizes=(20, 200),  # Control the size of the bubbles (min and max sizes)
                alpha=0.6, 
                palette='coolwarm',  # Choose a color palette for better visualization
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Age vs. Frequency of Participation with Job Satisfaction')
plt.xlabel('Age')
plt.ylabel('Frequency of Participation ')

# Add a legend for Job Satisfaction and size of bubbles
plt.legend(title='Job Satisfaction', loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here

# Clean the data by removing NaN values in the relevant columns
df_cleaned_compensation = df.dropna(subset=['ConvertedCompYearly', 'JobSat', 'Age'])

# Normalize the size of bubbles to reflect the actual Age range (optional)
bubble_size = df_cleaned_compensation['Age'] * 5  # Scale Age (to control bubble size)

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=df_cleaned_compensation,
                x='ConvertedCompYearly', 
                y='JobSat', 
                size=bubble_size,
                sizes=(20, 200),
                alpha=0.6, 
                palette='coolwarm',
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Bubble Plot: Yearly Compensation vs. Job Satisfaction with Respondent Age')
plt.xlabel('Yearly Compensation (ConvertedCompYearly)')
plt.ylabel('Job Satisfaction (JobSat)')

# Add a custom legend for bubble sizes (representing Age)
plt.legend(title='Age', loc='upper right')

plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
# Print the unique values of the 'LanguageHaveWorkedWith' column
unique_languages = df['LanguageHaveWorkedWith'].unique()
print("Unique values of LanguageHaveWorkedWith:")
print(unique_languages)

# Print the sum of the 'LanguageHaveWorkedWith' column (i.e., the total number of responses)
language_counts_sum = df['LanguageHaveWorkedWith'].count()
print(f"\nTotal count of 'LanguageHaveWorkedWith' responses: {language_counts_sum}")
```

```python
# Remove NaN values and extract the first language before the semicolon
df_cleaned_languages = df.dropna(subset=['LanguageHaveWorkedWith'])

# Extract the first language before the semicolon
df_cleaned_languages['PrimaryLanguage'] = df_cleaned_languages['LanguageHaveWorkedWith'].str.split(';').str[0]

# Print the unique values of the new 'PrimaryLanguage' column
unique_primary_languages = df_cleaned_languages['PrimaryLanguage'].unique()
print("Unique primary languages (first language before semicolon):")
print(unique_primary_languages)

# Print the count of each unique primary language
language_counts = df_cleaned_languages['PrimaryLanguage'].value_counts()
print("\nCount of each primary language:")
print(language_counts)
```

```python
##Write your code here

# Extract the first language before the semicolon
df_cleaned_languages['PrimaryLanguage'] = df_cleaned_languages['LanguageHaveWorkedWith'].str.split(';').str[0]

# Count the frequency of each language for each age group
language_counts = df_cleaned_languages.groupby(['Age', 'PrimaryLanguage']).size().reset_index(name='LanguageFreq')

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=language_counts,
                x='Age', 
                y='PrimaryLanguage', 
                size='LanguageFreq', 
                sizes=(20, 200), 
                alpha=0.6, 
                palette='coolwarm', 
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Technology Preferences by Age with Frequency of Languages')
plt.xlabel('Age')
plt.ylabel('Programming Language (Primary Language)')

# Show the plot
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
# Clean the data by removing NaN values in the relevant columns
df_cleaned_databases = df.dropna(subset=['JobSat', 'DatabaseWantToWorkWith'])

# Extract the first database before the semicolon
df_cleaned_databases['PrimaryDatabase'] = df_cleaned_databases['DatabaseWantToWorkWith'].str.split(';').str[0]

# Count the frequency of each database for each job satisfaction level
database_counts = df_cleaned_databases.groupby(['JobSat', 'PrimaryDatabase']).size().reset_index(name='DatabaseFreq')
```

```python
##Write your code here

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=database_counts,
                x='JobSat', 
                y='PrimaryDatabase', 
                size='DatabaseFreq',  # Size of bubbles represents the frequency of databases
                sizes=(20, 200),  # Control the size of the bubbles (min and max sizes)
                alpha=0.6, 
                palette='coolwarm',  # Choose a color palette for better visualization
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Preferred Databases vs. Job Satisfaction')
plt.xlabel('Job Satisfaction ')
plt.ylabel('Preferred Database')

# Add a legend and position it outside the chart
plt.legend(title='Job Satisfaction', loc='upper left', bbox_to_anchor=(1.05, 1))

# Show the plot
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here

# Clean the data by removing NaN values in the relevant columns
df_cleaned_compensation_roles = df.dropna(subset=['ConvertedCompYearly', 'DevType', 'JobSat'])

# Group by Developer Role (DevType) and get the average JobSat for each role
role_compensation = df_cleaned_compensation_roles.groupby('DevType').agg(
    avg_compensation=('ConvertedCompYearly', 'mean'),
    avg_job_satisfaction=('JobSat', 'mean'),
    num_respondents=('ConvertedCompYearly', 'size')
).reset_index()

# Normalize the JobSat values for bubble size
min_size = 20 
max_size = 200
role_compensation['bubble_size'] = min_size + (role_compensation['avg_job_satisfaction'] - 1) * (max_size - min_size) / 9

# Create the bubble plot
plt.figure(figsize=(14, 8))

scatter = sns.scatterplot(data=role_compensation,
                          x='avg_compensation', 
                          y='DevType', 
                          size='bubble_size',
                          sizes=(min_size, max_size),
                          alpha=0.6, 
                          hue='avg_job_satisfaction',
                          palette='coolwarm',
                          edgecolor='w', 
                          linewidth=0.5)

# Adding titles and labels
plt.title('Bubble Plot: Compensation Across Developer Roles with Job Satisfaction')
plt.xlabel('Average Yearly Compensation')
plt.ylabel('Developer Role')

# Add a color legend for job satisfaction
plt.legend(title='Job Satisfaction', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here

# Clean the data by removing NaN values in the relevant columns
df_cleaned_collab_tools = df.dropna(subset=['Age', 'NEWCollabToolsHaveWorkedWith'])

# Extract the first tool in the 'NEWCollabToolsHaveWorkedWith' column
df_cleaned_collab_tools['PrimaryCollabTool'] = df_cleaned_collab_tools['NEWCollabToolsHaveWorkedWith'].str.split(';').str[0]

# Count the frequency of each collaboration tool usage by age group
collab_tool_counts = df_cleaned_collab_tools.groupby(['Age', 'PrimaryCollabTool']).size().reset_index(name='ToolUsageFreq')

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=collab_tool_counts,
                x='Age', 
                y='PrimaryCollabTool', 
                size='ToolUsageFreq',
                sizes=(20, 200),  
                alpha=0.6, 
                palette='coolwarm',
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Collaboration Tools Usage by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Tool')

plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here

# Clean the data by removing NaN values in the relevant columns
df_cleaned_web_frameworks = df.dropna(subset=['JobSat', 'WebframeWantToWorkWith'])

# Extract the first framework in the 'WebframeWantToWorkWith' column
df_cleaned_web_frameworks['PrimaryWebFramework'] = df_cleaned_web_frameworks['WebframeWantToWorkWith'].str.split(';').str[0]

# Count the frequency of each web framework by job satisfaction
web_framework_counts = df_cleaned_web_frameworks.groupby(['JobSat', 'PrimaryWebFramework']).size().reset_index(name='FrameworkUsageFreq')

# Create the bubble plot
plt.figure(figsize=(14, 8))

sns.scatterplot(data=web_framework_counts,
                x='JobSat', 
                y='PrimaryWebFramework', 
                size='FrameworkUsageFreq',
                sizes=(20, 200),  
                alpha=0.6, 
                palette='coolwarm',
                edgecolor='w', 
                linewidth=0.5)

# Adding titles and labels
plt.title('Preferred Web Frameworks vs. Job Satisfaction')
plt.xlabel('Job Satisfaction ')
plt.ylabel('Preferred Web Framework')

plt.tight_layout()
plt.show()
```
