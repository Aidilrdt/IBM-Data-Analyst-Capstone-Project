```python
from scipy.stats import spearmanr, pearsonr
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here

# Column names
print("Column Names:")
print(df.columns)

# data types
print("\nData Types:")
print(df.dtypes)

# summary information
print("\nSummary Information:")
print(df.info())

# Statistics
print("\Statistics:")
print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here

# Impute missing numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Impute missing categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

print(df.isnull().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here

# Employment column value counts
print("Employment Value Counts:")
print(df['Employment'].value_counts(), "\n")

# JobSat column value counts
print("JobSat (Job Satisfaction) Value Counts:")
print(df['JobSat'].value_counts(), "\n")

# YearsCodePro column value counts
print("YearsCodePro (Professional Coding Experience) Value Counts:")
print(df['YearsCodePro'].value_counts(), "\n")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here

job_sat = df['JobSat']

# Create a KDE plot for JobSat (converted to numerical representation)
job_sat_numeric = job_sat.replace({
    "Very satisfied": 5,
    "Satisfied": 4,
    "Neither satisfied nor dissatisfied": 3,
    "Dissatisfied": 2,
    "Very dissatisfied": 1
})

plt.figure(figsize=(8, 6))
sns.kdeplot(job_sat_numeric, shade=True, color='blue')
plt.title("Job Satisfaction Distribution", fontsize=14)
plt.xlabel("Job Satisfaction (1=Very Dissatisfied, 5=Very Satisfied)")
plt.ylabel("Density")
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here

# Split the languages into individual rows for both columns
worked_split = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
want_split = df['LanguageWantToWorkWith'].dropna().str.split(';').explode()

# Calculate the frequency of each language
worked_counts = worked_split.value_counts()
want_counts = want_split.value_counts()

# Combine the frequencies into a DataFrame
languages_comparison = pd.DataFrame({
    'HaveWorkedWith': worked_counts,
    'WantToWorkWith': want_counts
}).fillna(0).astype(int)

# Sort by the total count of both columns
languages_comparison['Total'] = languages_comparison['HaveWorkedWith'] + languages_comparison['WantToWorkWith']
languages_comparison = languages_comparison.sort_values(by='Total', ascending=False)

# Display the comparison
print("Programming Language Frequency Comparison:")
print(languages_comparison)
```

```python
# Extract count of languages
worked_languages = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode().value_counts()
want_languages = df['LanguageWantToWorkWith'].dropna().str.split(';').explode().value_counts()

# Convert the counts into DataFrames
worked_df = pd.DataFrame({'Language': worked_languages.index, 'HaveWorkedWith': worked_languages.values})
want_df = pd.DataFrame({'Language': want_languages.index, 'WantToWorkWith': want_languages.values})

# Merge the DataFrames
merged_df = pd.merge(worked_df, want_df, on='Language', how='outer').fillna(0)

# Plot the bar chart
plt.figure(figsize=(14, 6))
merged_df.set_index('Language').plot(kind='bar', stacked=False, color=['blue', 'orange'])
plt.title('Comparison of Programming Languages: Worked With vs Want to Work With', fontsize=14)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here

# Define regions and their corresponding countries
region_mapping = {
    'Africa': [
        'Algeria', 'Nigeria', 'South Africa', 'Egypt', 'Kenya', 'Ethiopia',
        'Morocco', 'Tunisia', 'Uganda', 'Tanzania', 'Ghana', 'Cameroon',
        'Zimbabwe', 'Senegal', 'Sudan', 'Zambia', 'Botswana', 'Rwanda'
    ],
    'Asia': [
        'India', 'China', 'Japan', 'Malaysia', 'Singapore', 'Pakistan',
        'Indonesia', 'Philippines', 'Thailand', 'Bangladesh', 'Vietnam',
        'Sri Lanka', 'Saudi Arabia', 'United Arab Emirates', 'Iran, Islamic Republic of...',
        'Korea, Republic of', 'Israel'
    ],
    'Europe': [
        'United Kingdom of Great Britain and Northern Ireland', 'Germany', 'France',
        'Italy', 'Spain', 'Netherlands', 'Switzerland', 'Sweden', 'Norway',
        'Poland', 'Ireland', 'Austria', 'Belgium', 'Finland', 'Denmark',
        'Portugal', 'Ukraine', 'Czech Republic', 'Romania', 'Greece', 'Russia'
    ],
    'North America': [
        'United States of America', 'Canada', 'Mexico'
    ],
    'South America': [
        'Brazil', 'Argentina', 'Chile', 'Colombia', 'Peru', 'Ecuador', 'Venezuela, Bolivarian Republic of...'
    ],
    'Oceania': [
        'Australia', 'New Zealand', 'Fiji', 'Papua New Guinea'
    ],
    'Nomadic': ['Nomadic']
}

# Function to map a country to its region
def map_country_to_region_manual(country):
    for region, countries in region_mapping.items():
        if country in countries:
            return region
    return 'Nomadic'

# Map the countries in the dataset
df['Region'] = df['Country'].apply(map_country_to_region_manual)

# Group by Region and RemoteWork
region_remotework = df.groupby(['Region', 'RemoteWork']).size().unstack()

# Plot the grouped bar chart
plt.figure(figsize=(12, 6))
region_remotework.plot(kind='bar', figsize=(14, 8), stacked=False, colormap='viridis')

plt.title('Distribution of Remote Work by Region', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Remote Work Type', loc='upper right')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
# Map non-numerical values to numeric equivalents in YearsCodePro
years_code_mapping = {
    "Less than 1 year": 0,
    "More than 50 years": 51
}

# Replace the non-numerical values in YearsCodePro
df['YearsCodePro'] = df['YearsCodePro'].replace(years_code_mapping)

# Convert the column to numeric (handling any remaining invalid values)
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Remove NaN values from YearsCodePro
df = df.dropna(subset=['YearsCodePro'])

# Remove NaN values from JobSat column
df = df.dropna(subset=['JobSat'])

print("YearsCodePro:")
print(df['YearsCodePro'].unique())

print("JobSat:")
print(df['JobSat'].unique())
```

```python
## Write your code here

# Map non-numerical values to numeric equivalents for JobSat
job_sat_mapping = {
    "Very dissatisfied": 1,
    "Dissatisfied": 2,
    "Neither satisfied nor dissatisfied": 3,
    "Satisfied": 4,
    "Very satisfied": 5
}

# Replace non-numerical values in JobSat column
df['JobSat'] = df['JobSat'].replace(job_sat_mapping)

# Calculate Spearman correlation
correlation, p_value = spearmanr(df['JobSat'], df['YearsCodePro'])

# Display the correlation coefficient and p-value
print(f"Spearman correlation between JobSat and YearsCodePro: {correlation}")
print(f"P-value: {p_value}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
# Split the 'Employment' column by ';' and create a new dataframe for the categories
employment_split = df['Employment'].str.split(';', expand=True)

# Define the categories for employment statuses
categories = {
    'Employed': ['Employed, full-time', 'Employed, part-time'],
    'Student': ['Student, full-time', 'Student, part-time'],
    'Independent Contractor / Freelancer / Self-employed': ['Independent contractor, freelancer, or self-employed'],
    'Not Employed': ['Not employed, but looking for work', 'Not employed, and not looking for work'],
    'Retired': ['Retired'],
    'Prefer Not to Say': ['I prefer not to say']
}

# Function to map employment status to categories
def categorize_employment(status):
    for category, keywords in categories.items():
        if any(keyword in status for keyword in keywords):
            return category
    return 'Other'  # In case any status doesn't match

# Reshape the split employment data to a single column
employment_melted = employment_split.melt(value_name='Employment').dropna(subset=['Employment'])

# Apply categorization
employment_melted['Category'] = employment_melted['Employment'].apply(categorize_employment)
```

```python
## Write your code here

# Create a cross-tabulation between employment categories and education levels
employment_education_crosstab = pd.crosstab(employment_melted['Category'], df['EdLevel'], margins=True, margins_name="Total")

# Display the cross-tabulation
print(employment_education_crosstab) 

# Plot a stacked bar chart
employment_education_crosstab.drop('Total', axis=1).T.plot(kind='bar', stacked=True, figsize=(24, 18), colormap='Set2')

# Add labels and title
plt.title('Relationship Between Employment Status and Education Level', fontsize=14)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
